```
import numpy as np

x = np.linspace(0,5,11)
y= x ** 2
```

```
x
```

```
y
```

```
#functional methos
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
#plt.show() # if not in the jupyter notebook
```

```
#multiplot
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```
#object oriented method
fig = plt.figure() # create fig object

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # args left, bottom, width,height
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```
fig = plt.figure()

axis1= fig.add_axes([0.1,0.1,0.8,0.8])
axis2= fig.add_axes([0.2,0.5,0.4,0.3]) # 20 % from left 50 % from bottom 
#40% of canvas width and 30 % of canvas height.
#the number must be between 0 and 1
axis1.plot(x,y)

axis2.plot(y,x)
```

```
fig,axis = plt.subplots(nrows=3,ncols=3) # in subplot you can spicify the number of roww and column
# axis.plot(x,y)
plt.tight_layout() # fix the overlap in the layout
```

```
fig, axis = plt.subplots(nrows=1,ncols=2)
axis[0].plot(x,y)
axis[0].set_title("Title")
axis[1].plot(y,x)
axis[1].set_title("title2")
```

```
for current_ax in axis:
  current_ax.plot(x,y)
```

```
axis
```

```
# Figure size and dpi
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```
fig =plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
ax.plot(x,x**2,label='X square')
ax.plot(x,x**3,label='x cube')
ax.legend(loc=0)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')

```

```
#matplotlib exercise
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```
import matplotlib.pyplot as plt
%matplotlib inline # tw the show the figure
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

```
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y,color='red')
ax2.plot(x,y)
```

```
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

```
fig,axes = plt.subplots(1,2,figsize=(12,2))

axes[0].plot(x,y,lw=3,ls='--')
axes[1].plot(x,z,lw=3,color='red')
```
